## Level 1 – Task 1: Data Exploration and Preprocessing

The objective of this task is to explore the dataset, understand its structure, handle missing values, and analyze the distribution of the target variable. This step ensures that the dataset is clean and ready for further analysis.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("Dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
df.fillna("Unknown", inplace=True)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Aggregate rating"], bins=10, kde=True)
plt.show()
```

## Level 1 – Task 2: Descriptive Analysis

This task focuses on summarizing the dataset using statistical measures and analyzing the distribution of categorical variables such as cities, countries, and cuisines.


```python
df.describe()
```

```python
df["Country Code"].value_counts().head(10)
```

```python
df["City"].value_counts().head(10)
```

```python
df["Cuisines"].value_counts().head(10)
```

```python
top_cities = df["City"].value_counts().head(10)

plt.figure(figsize=(10,5))
top_cities.plot(kind="bar")
plt.show()
```

```python
all_cuisines = df["Cuisines"].str.split(", ").explode()
top_cuisines = all_cuisines.value_counts().head(10)

plt.figure(figsize=(10,5))
top_cuisines.plot(kind="bar")
plt.show()
```

## Level 1 – Task 3: Geospatial Analysis

This task focuses on analyzing the geographical distribution of restaurants using latitude and longitude and examining whether location influences restaurant ratings.


```python
plt.figure(figsize=(10,6))
plt.scatter(df["Longitude"], df["Latitude"], alpha=0.5)
plt.title("Geographical Distribution of Restaurants")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
df["Country Code"].value_counts().head(10)
```

```python
top_countries = df["Country Code"].value_counts().head(10)

plt.figure(figsize=(10,5))
top_countries.plot(kind="bar")
plt.title("Top Countries by Number of Restaurants")
plt.xlabel("Country Code")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(
    df["Longitude"],
    df["Latitude"],
    c=df["Aggregate rating"],
    cmap="viridis",
    alpha=0.6
)
plt.colorbar(label="Aggregate Rating")
plt.title("Restaurant Ratings by Location")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

### Geospatial Insights

- Restaurants are concentrated in specific geographical regions, particularly urban areas.
- Certain locations show clusters of higher-rated restaurants.
- While location influences restaurant density, ratings are affected by multiple factors beyond geography.
